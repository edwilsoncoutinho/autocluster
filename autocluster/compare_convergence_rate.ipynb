```python
def plot_convergence(min_cost_ls):
    # plotting
    plt.figure(figsize=(10,10))
    plt.plot(min_cost_ls, linestyle='-', marker='o', color='b')
    plt.xlabel('n_evaluations', color='white', fontsize=15)
    plt.ylabel('performance of best configuration', color='white', fontsize=15)
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
    plt.show()
```

```python
dataset = 'dim7.csv'
```

```python
plot_convergence(d1[dataset]['convergence_curve'])
plot_convergence(d2[dataset]['convergence_curve'])
plot_convergence(d3[dataset]['convergence_curve'])
```
