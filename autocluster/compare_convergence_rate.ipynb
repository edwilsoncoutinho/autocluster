```python
def plot_convergence_curve(min_cost_ls):
    # plotting
    plt.figure(figsize=(10,10))
    plt.plot(min_cost_ls, linestyle='-', marker='o', color='b')
    plt.xlabel('n_evaluations', color='white', fontsize=15)
    plt.ylabel('performance of best configuration', color='white', fontsize=15)
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
    plt.show()
```

```python
dataset = 'dim14.csv'
```

```python
plot_convergence_curve(d1[dataset]['convergence_curve'])
plot_convergence_curve(d2[dataset]['convergence_curve'])
plot_convergence_curve(d3[dataset]['convergence_curve'])
plot_convergence_curve(d4[dataset]['convergence_curve'])
```

```python
plot_params = {
    'data': [
#         d1[dataset]['convergence_curve'], 
        d2[dataset]['convergence_curve'], 
        d3[dataset]['convergence_curve'], 
        d4[dataset]['convergence_curve']
    ],
    'iterations': 'max',
    'title': 'Convergence rates of difference approaches',
    'legends': ['SMAC only', 'SMAC + Warmstart', 'Random Optimization']
}

plot_convergence(**plot_params)
```
